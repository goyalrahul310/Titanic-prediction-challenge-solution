```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier # pending
from sklearn.metrics import accuracy_score # pend
```

```python
#loading the file
data = pd.read_csv("Train.csv")
data.head(n =15)
```

```python
data.info()
```

```python
# removing column which are not required
columns_to_drop = ["name","ticket","cabin","embarked","boat","body","home.dest"]
data_clean = data.drop(columns_to_drop,axis = 1)
```

```python
data_clean.head()
```

```python
data_clean.info()
```

```python
# transforming strings into intergers
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data_clean["sex"] = le.fit_transform(data_clean["sex"])
```

```python
data_clean.head()
```

```python
# filling columm of age which are not known
data_clean = data_clean.fillna(data_clean["age"].mean())
data_clean.info()
```

```python
input_column  = ["pclass","sex","age","sibsp","parch","fare"]
output_column = ["survived"]

X = data_clean[input_column]
Y = data_clean[output_column]
print(X.shape,Y.shape)
print(X.head())
```

```python
model = RandomForestClassifier(n_estimators = 100)
model.fit(X,Y)
```

```python
# train accuracy
print("train accuracy : ",accuracy_score(Y,model.predict(X)))
```
